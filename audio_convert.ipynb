```python
import os
def do_files(f=""):
    files = os.listdir(f"data/audio_data{f}/")
    if "!ids.txt" in files:
        files.remove("!ids.txt")
    if "!info.txt" in files:
        files.remove("!info.txt")
    files = set([int(i.split(".")[0]) for i in files])
    return files
```

```python
# Создание NPY
import time
start_time = time.time()

import librosa
import numpy as np

max_audio = 260142
exclude_files = [3591, 3979, 4469, 5184, 5185, 5186, 5187, 5188, 5189, 5190, 5191, 54973]
existing_mp3_files = do_files()
for i in exclude_files:
    if i in existing_mp3_files:
        existing_mp3_files.remove(i)
existing_stft_files = do_files(f="_stft")

errored = []
proc = 0
for i in existing_mp3_files:
    if i % 1000 == 1:
        new_time = time.time() - start_time
        print("Process:", i, f"{proc/new_time}")
    if i not in existing_stft_files:
        proc += 1
        try:
            y, sr = librosa.load(f'data/audio_data/{i}.mp3', sr=11025)
            D_transpose = np.transpose(librosa.stft(y, n_fft=1024))
            np.save(f'data/audio_data_stft/{i}.npy', D_transpose)
        except:
            errored.append(i)
            print("Errored:", i)
    
if len(errored) != 0:
    print(errored)
```
