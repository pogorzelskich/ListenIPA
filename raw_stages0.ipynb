```python
common_list = set()
# languages_count = dict()
for raw_line in all_raw_lines:
    line = json.loads(raw_line)
    new_line = {}
    curr_keys = line.keys()
    if "sounds" in curr_keys:
        sounds = line["sounds"]
        having_ipa = False
        having_audio = False
        for sound in sounds:
            if "audio" in sound:
                having_ipa = True
            if "ipa" in sound:
                having_audio = True
        
        if having_ipa and having_audio:
            lang_str = line["lang"] + "\t" + line["lang_code"]
            new_line["l"] = lang_str
            if "word" in curr_keys:
                new_line["w"] = line["word"]
            new_line["s"] = line["sounds"]
            common_list.add(json.dumps(new_line, ensure_ascii=False))
```
